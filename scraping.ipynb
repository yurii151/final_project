```python
from bs4 import BeautifulSoup
import requests 
from urllib import request
import os
import pandas as pd
from requests_html import HTMLSession
import urllib.request
import re
import time
from urllib.error import URLError
from urllib.parse import quote
```

```python
books_df = pd.read_csv("Resources/Nilim-Kaggle/books.csv", error_bad_lines=False)
mylist = books_df['title'].tolist()
print(mylist)
```

```python
df_genre = pd.read_csv('Resources/genre.csv')
list_of_genre = df_genre['genres'].tolist()
print(list_of_genre)
```

# For Loop Code for looping through full list of books. Need to add code to catch errors.

```python
for query in mylist:
    try:
        time.sleep(.1)
        start_url = "https://google.com/search?q=" + query + ' book'
        url = start_url.replace(" ","+")
        #print(url)
        #using a User-Agent to make the search
        req = urllib.request.Request(url, headers={'User-Agent' : "Magic Browser"})
        #opening URL
        r = requests.get(url)
        response = urllib.request.urlopen( req )
        html = response.read()
        #htmlstr = bytecode.decode()
        soup = BeautifulSoup(r.text,features="lxml")
        text= soup.get_text().format('left aligned','{},{}','{},{}')
        #print(text)

        #find_genre =text[text.find('Genre: ')+7:text.find('Genre: ')+20] 
        for genre in list_of_genre:
                if re.search(genre, text, re.IGNORECASE):
                    #print(query+"="+genre)      
                    books_df_with_genre = books_df.append({'Genre': genre}, ignore_index=True)
    except:
        print("Error on genre "+genre+" query "+query)
            
print(books_df_with_genre)

books_df_with_genre.to_csv("books_df_with_genre.csv")
```
